```python
import numpy as np
import math
```

```python
# Define useful constants
mt=172.5*10**-3 #top mark mass in TeV
v=246*10**-3 #vacuum expectation of the Higgs in TeV
mw=80.38*10**-3 #W boson mass in TeV
```

```python
# List the constraints on all the coefficients
Lam_mmtt_LR_neg=1.5
Lam_mmtt_LR_pos=1.5
Lam_mmtt_LR_max=1.5

Lam_mmtt_RR_neg=1.4
Lam_mmtt_RR_pos=1.4
Lam_mmtt_RR_max=1.4

Lam_mmcc_LR_neg=2.1
Lam_mmcc_LR_pos=2.0
Lam_mmcc_LR_max=2.1

Lam_mmcc_RR_neg=2.1
Lam_mmcc_RR_pos=2.0
Lam_mmcc_RR_max=2.1

Lam_mmuu_LR_neg=8.0
Lam_mmuu_LR_pos=5.9
Lam_mmuu_LR_max=8.0

Lam_mmuu_RR_neg=9.2
Lam_mmuu_RR_pos=5.1
Lam_mmuu_RR_max=9.2

Lam_eett_LR_neg=1.6
Lam_eett_LR_pos=2.3
Lam_eett_LR_max=2.3

Lam_eett_RR_neg=2.1
Lam_eett_RR_pos=2.5
Lam_eett_RR_max=2.5

Lam_eecc_LR_neg=2.0
Lam_eecc_LR_pos=2.0
Lam_eecc_LR_max=2.0

Lam_eecc_RR_neg=2.0
Lam_eecc_RR_pos=2.0
Lam_eecc_RR_max=2.0

Lam_eeuu_LR_neg=8.0
Lam_eeuu_LR_pos=6.5
Lam_eeuu_LR_max=8.0

Lam_eeuu_RR_neg=8.8
Lam_eeuu_RR_pos=5.9
Lam_eeuu_RR_max=8.8

# values from top quark experiment paper
Lam_mmtt_LR_neg_pap=.63
Lam_mmtt_LR_pos_pap=.58
Lam_mmtt_LR_max_pap=.63

Lam_eett_LR_neg_pap=.63
Lam_eett_LR_pos_pap=.58
Lam_eett_LR_max_pap=.63

Lam_mmtt_RR_neg_pap=.63
Lam_mmtt_RR_pos_pap=.56
Lam_mmtt_RR_max_pap=.63

Lam_eett_RR_neg_pap=.63
Lam_eett_RR_pos_pap=.56
Lam_eett_RR_max_pap=.63
```

```python
# function to calculate \Delta F=2 coefficients

def C_mnpq(Lam_mmpp,Lam_mmqq,Lam_nnpp,Lam_nnqq):
    C_mmpp=pow(Lam_mmpp,-2)
    C_mmqq=pow(Lam_mmqq,-2)
    C_nnpp=pow(Lam_nnpp,-2)
    C_nnqq=pow(Lam_nnqq,-2)
    C=np.sqrt(C_mmpp*C_nnqq)+np.sqrt(C_mmqq*C_nnpp)+4*pow(C_mmpp*C_mmqq*C_nnpp*C_nnqq,.25)
    return C
```

```python
# function to calculate \Delta F=2 coefficients assuming \Delta F=1 coef are 0
def C_mnpq_F10(Lam_mmpp,Lam_mmqq,Lam_nnpp,Lam_nnqq):
    C_mmpp=1/(Lam_mmpp)**2
    C_mmqq=1/(Lam_mmqq)**2
    C_nnpp=1/(Lam_nnpp)**2
    C_nnqq=1/(Lam_nnqq)**2
    C=np.sqrt(C_mmpp*C_nnqq)+np.sqrt(C_mmqq*C_nnpp)
    return C
```

```python
# Function to calculate branching Ratio

def BR(Lam_mmpp_LR,Lam_mmqq_LR,Lam_nnpp_LR,Lam_nnqq_LR,Lam_mmpp_RR,Lam_mmqq_RR,Lam_nnpp_RR,Lam_nnqq_RR):
    C_mnpq_LR=C_mnpq(Lam_mmpp_LR,Lam_mmqq_LR,Lam_nnpp_LR,Lam_nnqq_LR)
    C_mnpq_RR=C_mnpq(Lam_mmpp_RR,Lam_mmqq_RR,Lam_nnpp_RR,Lam_nnqq_RR)
#     Switching the n,m doesn't change the value of the formula
    B=pow(96*math.pi**2,-1)*mt**2*v**2*(2*C_mnpq_LR**2+2*C_mnpq_RR**2)*pow(1-mw**2/mt**2,-2)*pow(1+2*mw**2/mt**2,-1)
    return B
    
```

```python
# Function to calculate branching ratio assumung \Delta F=1 coef are 0
def BR_F10(Lam_mmpp_LR,Lam_mmqq_LR,Lam_nnpp_LR,Lam_nnqq_LR,Lam_mmpp_RR,Lam_mmqq_RR,Lam_nnpp_RR,Lam_nnqq_RR):
    C_mnpq_LR=C_mnpq_F10(Lam_mmpp_LR,Lam_mmqq_LR,Lam_nnpp_LR,Lam_nnqq_LR)
    C_mnpq_RR=C_mnpq_F10(Lam_mmpp_RR,Lam_mmqq_RR,Lam_nnpp_RR,Lam_nnqq_RR)
#     Switching the n,m doesn't change the value of the formula
    B=pow(96*math.pi**2,-1)*mt**2*v**2*(2*C_mnpq_LR**2+2*C_mnpq_RR**2)*pow(1-mw**2/mt**2,-2)*pow(1+2*mw**2/mt**2,-1)
    return B
```

```python
# Branching Ratios with initial Wolfgang values
BR_tcme_neg=BR(Lam_eett_LR_neg,Lam_mmtt_LR_neg,Lam_eecc_LR_neg,Lam_mmcc_LR_neg,
               Lam_eett_RR_neg,Lam_mmtt_RR_neg,Lam_eecc_RR_neg,Lam_mmcc_RR_neg)
BR_tcme_pos=BR(Lam_eett_LR_pos,Lam_mmtt_LR_pos,Lam_eecc_LR_pos,Lam_mmcc_LR_pos,
               Lam_eett_RR_pos,Lam_mmtt_RR_pos,Lam_eecc_RR_pos,Lam_mmcc_RR_pos)
BR_tcme_max=BR(Lam_eett_LR_max,Lam_mmtt_LR_max,Lam_eecc_LR_max,Lam_mmcc_LR_max,
               Lam_eett_RR_max,Lam_mmtt_RR_max,Lam_eecc_RR_max,Lam_mmcc_RR_max)

BR_tume_neg=BR(Lam_eett_LR_neg,Lam_mmtt_LR_neg,Lam_eeuu_LR_neg,Lam_mmuu_LR_neg,
               Lam_eett_RR_neg,Lam_mmtt_RR_neg,Lam_eeuu_RR_neg,Lam_mmuu_RR_neg)
BR_tume_pos=BR(Lam_eett_LR_pos,Lam_mmtt_LR_pos,Lam_eeuu_LR_pos,Lam_mmuu_LR_pos,
               Lam_eett_RR_pos,Lam_mmtt_RR_pos,Lam_eeuu_RR_pos,Lam_mmuu_RR_pos)
BR_tume_max=BR(Lam_eett_LR_max,Lam_mmtt_LR_max,Lam_eeuu_LR_max,Lam_mmuu_LR_max,
               Lam_eett_RR_max,Lam_mmtt_RR_max,Lam_eeuu_RR_max,Lam_mmuu_RR_max)

# Branching Ratios with exp t coef included
BR_tcme_neg_pap=BR(Lam_eett_LR_neg_pap,Lam_mmtt_LR_neg_pap,Lam_eecc_LR_neg,Lam_mmcc_LR_neg,
               Lam_eett_RR_neg_pap,Lam_mmtt_RR_neg_pap,Lam_eecc_RR_neg,Lam_mmcc_RR_neg)
BR_tcme_pos_pap=BR(Lam_eett_LR_pos_pap,Lam_mmtt_LR_pos_pap,Lam_eecc_LR_pos,Lam_mmcc_LR_pos,
               Lam_eett_RR_pos_pap,Lam_mmtt_RR_pos_pap,Lam_eecc_RR_pos,Lam_mmcc_RR_pos)
BR_tcme_max_pap=BR(Lam_eett_LR_max_pap,Lam_mmtt_LR_max_pap,Lam_eecc_LR_max,Lam_mmcc_LR_max,
               Lam_eett_RR_max_pap,Lam_mmtt_RR_max_pap,Lam_eecc_RR_max,Lam_mmcc_RR_max)

BR_tume_neg_pap=BR(Lam_eett_LR_neg_pap,Lam_mmtt_LR_neg_pap,Lam_eeuu_LR_neg,Lam_mmuu_LR_neg,
               Lam_eett_RR_neg_pap,Lam_mmtt_RR_neg_pap,Lam_eeuu_RR_neg,Lam_mmuu_RR_neg)
BR_tume_pos_pap=BR(Lam_eett_LR_pos_pap,Lam_mmtt_LR_pos_pap,Lam_eeuu_LR_pos,Lam_mmuu_LR_pos,
               Lam_eett_RR_pos_pap,Lam_mmtt_RR_pos_pap,Lam_eeuu_RR_pos,Lam_mmuu_RR_pos)
BR_tume_max_pap=BR(Lam_eett_LR_max_pap,Lam_mmtt_LR_max_pap,Lam_eeuu_LR_max,Lam_mmuu_LR_max,
               Lam_eett_RR_max_pap,Lam_mmtt_RR_max_pap,Lam_eeuu_RR_max,Lam_mmuu_RR_max)

# Branching Ratios with initial Wolfgang values with assumption \Delta F=1 coef are 0
BR_tcme_neg_F10=BR_F10(Lam_eett_LR_neg,Lam_mmtt_LR_neg,Lam_eecc_LR_neg,Lam_mmcc_LR_neg,
               Lam_eett_RR_neg,Lam_mmtt_RR_neg,Lam_eecc_RR_neg,Lam_mmcc_RR_neg)
BR_tcme_pos_F10=BR_F10(Lam_eett_LR_pos,Lam_mmtt_LR_pos,Lam_eecc_LR_pos,Lam_mmcc_LR_pos,
               Lam_eett_RR_pos,Lam_mmtt_RR_pos,Lam_eecc_RR_pos,Lam_mmcc_RR_pos)
BR_tcme_max_F10=BR_F10(Lam_eett_LR_max,Lam_mmtt_LR_max,Lam_eecc_LR_max,Lam_mmcc_LR_max,
               Lam_eett_RR_max,Lam_mmtt_RR_max,Lam_eecc_RR_max,Lam_mmcc_RR_max)

BR_tume_neg_F10=BR_F10(Lam_eett_LR_neg,Lam_mmtt_LR_neg,Lam_eeuu_LR_neg,Lam_mmuu_LR_neg,
               Lam_eett_RR_neg,Lam_mmtt_RR_neg,Lam_eeuu_RR_neg,Lam_mmuu_RR_neg)
BR_tume_pos_F10=BR_F10(Lam_eett_LR_pos,Lam_mmtt_LR_pos,Lam_eeuu_LR_pos,Lam_mmuu_LR_pos,
               Lam_eett_RR_pos,Lam_mmtt_RR_pos,Lam_eeuu_RR_pos,Lam_mmuu_RR_pos)
BR_tume_max_F10=BR_F10(Lam_eett_LR_max,Lam_mmtt_LR_max,Lam_eeuu_LR_max,Lam_mmuu_LR_max,
               Lam_eett_RR_max,Lam_mmtt_RR_max,Lam_eeuu_RR_max,Lam_mmuu_RR_max)

# Branching Ratios with exp t coef included with assumption \Delta F=1 coef are 0
BR_tcme_neg_pap_F10=BR_F10(Lam_eett_LR_neg_pap,Lam_mmtt_LR_neg_pap,Lam_eecc_LR_neg,Lam_mmcc_LR_neg,
               Lam_eett_RR_neg_pap,Lam_mmtt_RR_neg_pap,Lam_eecc_RR_neg,Lam_mmcc_RR_neg)
BR_tcme_pos_pap_F10=BR_F10(Lam_eett_LR_pos_pap,Lam_mmtt_LR_pos_pap,Lam_eecc_LR_pos,Lam_mmcc_LR_pos,
               Lam_eett_RR_pos_pap,Lam_mmtt_RR_pos_pap,Lam_eecc_RR_pos,Lam_mmcc_RR_pos)
BR_tcme_max_pap_F10=BR_F10(Lam_eett_LR_max_pap,Lam_mmtt_LR_max_pap,Lam_eecc_LR_max,Lam_mmcc_LR_max,
               Lam_eett_RR_max_pap,Lam_mmtt_RR_max_pap,Lam_eecc_RR_max,Lam_mmcc_RR_max)

BR_tume_neg_pap_F10=BR_F10(Lam_eett_LR_neg_pap,Lam_mmtt_LR_neg_pap,Lam_eeuu_LR_neg,Lam_mmuu_LR_neg,
               Lam_eett_RR_neg_pap,Lam_mmtt_RR_neg_pap,Lam_eeuu_RR_neg,Lam_mmuu_RR_neg)
BR_tume_pos_pap_F10=BR_F10(Lam_eett_LR_pos_pap,Lam_mmtt_LR_pos_pap,Lam_eeuu_LR_pos,Lam_mmuu_LR_pos,
               Lam_eett_RR_pos_pap,Lam_mmtt_RR_pos_pap,Lam_eeuu_RR_pos,Lam_mmuu_RR_pos)
BR_tume_max_pap_F10=BR_F10(Lam_eett_LR_max_pap,Lam_mmtt_LR_max_pap,Lam_eeuu_LR_max,Lam_mmuu_LR_max,
               Lam_eett_RR_max_pap,Lam_mmtt_RR_max_pap,Lam_eeuu_RR_max,Lam_mmuu_RR_max)
```
